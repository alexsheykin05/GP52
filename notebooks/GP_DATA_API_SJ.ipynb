Теперь по циклу пройдемся по всем параметрам

```python
import time
```

```python
vacancies_list = []
```

```python
params = {'town': 4,
    'post_ids': 651,
    'experience': 1,
    'page': 1,
    'count': 100,
    'catalogues': 33}
request = requests.get('https://api.superjob.ru/2.0/vacancies/',
                        headers=headers, params=params)
vacs = request.json()
vacancy_category = vacancies_dict[651]
a = vacs.get('objects', [])
for i in a:
     i['vacancy_category'] = vacancy_category
a[0]
```

```python
from tqdm.auto import tqdm
```

```python
iters = len(town_ids) * len(vacancies_ids) * len(exp)
progress = tqdm(total=iters)
for t in town_ids:
    for v in vacancies_ids:
        vacancy_category = vacancies_dict[v]
        for e in exp:
            for page in range(20): #можно хоть 100, так как я настроил break
                params = {
                    'town': t,
                    'post_ids': v,
                    'experience': e,
                    'page': page,
                    'count': 100,
                    'catalogues': 33}
                try:
                    logging.info(f'отправлен запрос к апи вакансий')
                    request = requests.get('https://api.superjob.ru/2.0/vacancies/',
                                            headers=headers, params=params)
                    vacs = request.json()
                    vacs_list = vacs.get('objects', [])
                    if not vacs_list:
                        break
                    for i in vacs_list:
                        i['vacancy_category'] = vacancy_category
                        i['experience_level'] = e
                    vacancies_list += vacs_list
                    logging.info(f'успех для города {t}, категории вакансии {vacancy_category}, на странице {page}')
                except Exception as e:
                    logging.error(f'Error {e}')
                time.sleep(0.2)
            progress.update(1)
```

```python
df_superjob = pd.DataFrame(vacancies_list)
```

```python
df_superjob.to_csv('superjob.csv', encoding='utf-8-sig')
```
