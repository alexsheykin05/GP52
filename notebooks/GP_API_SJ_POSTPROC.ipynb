```python
cols_to_keep = ['vacancy_category', 'profession', 'payment_from', 'payment_to', 'currency',
                'candidat', 'vacancyRichText', 'town', 'experience',
                'type_of_work', 'place_of_work', 'firm_name']
```

```python
df = df_superjob[cols_to_keep]
```

```python
df
```

```python
import ast
```

```python
def get_title(col):
    col = ast.literal_eval(col)
    return col.get('title')
```

```python
for col in ['town', 'experience', 'type_of_work', 'place_of_work']:
     df[col] = df[col].apply(get_title)
```

```python
df
```

```python
df.drop(columns=['currency'], inplace=True)
```

```python
# Идея использования bs для очистки htmlвских артефактов

# Source - https://stackoverflow.com/a/12982689

# Posted by c24b, modified by community. See post 'Timeline' for change history

# Retrieved 2026-04-09, License - CC BY-SA 4.0

from bs4 import BeautifulSoup
```

```python
def clean_text(col):
    soup = BeautifulSoup(col, 'lxml')
    return soup.get_text(separator=' ').strip()
```

```python
df['vacancy_clean'] = df['vacancyRichText'].apply(clean_text)
df['candidat_clean'] = df['candidat'].apply(clean_text)
```

```python
df
```

vacancy_clean по итогу почистился лучше, так что оставим его

```python
df.drop(columns=['candidat_clean', 'vacancyRichText', 'candidat'], inplace=True)
```
